```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
sns.countplot(data=df, x="HiringDecision")
plt.title("Hiring Decision Distribution")
plt.show()
```

```python
sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution (0=Female, 1=Male)")
plt.show()
```

```python
sns.barplot(data=df, x="Gender", y="HiringDecision")
plt.title("Hiring Rate by Gender")
plt.ylabel("Hiring Probability")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop("HiringDecision", axis=1)
y = df["HiringDecision"]
```

```python
X_male = X[X["Gender"] == 1]
y_male = y[X["Gender"] == 1]
X_female = X[X["Gender"] == 0]
y_female = y[X["Gender"] == 0]
```

```python

X_train_male, X_test_male, y_train_male, y_test_male = train_test_split(X_male, y_male, test_size=0.2, random_state=42)
X_train_female, X_test_female, y_train_female, y_test_female = train_test_split(X_female, y_female, test_size=0.8, random_state=42)
```

```python
X_train = pd.concat([X_train_male, X_train_female])
y_train = pd.concat([y_train_male, y_train_female])
X_test = pd.concat([X_test_male, X_test_female])
y_test = pd.concat([y_test_male, y_test_female])
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

clf = LogisticRegression()
clf.fit(X_train_scaled, y_train)

y_pred = clf.predict(X_test_scaled)

```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from fairlearn.metrics import MetricFrame, selection_rate, demographic_parity_difference, equalized_odds_difference
```

```python
#gender as sensitive features 
sensitive_features = X_test["Gender"]

metric_frame = MetricFrame(
    metrics={"accuracy": accuracy_score, "selection_rate": selection_rate},
    y_true=y_test,
    y_pred=y_pred,
    sensitive_features=sensitive_features
)
```

```python
print("Metric Frame:")
print(metric_frame.by_group)

print("Demographic Parity Difference:", demographic_parity_difference(y_test, y_pred, sensitive_features=sensitive_features))
print("Equalized Odds Difference:", equalized_odds_difference(y_test, y_pred, sensitive_features=sensitive_features))
```

```python
from aif360.datasets import BinaryLabelDataset
from aif360.algorithms.preprocessing import Reweighing
```

```python
X["HiringDecision"] = y
data_aif = BinaryLabelDataset(
    df=X,
    label_names=["HiringDecision"],
    protected_attribute_names=["Gender"]
)
```

```python
RW = Reweighing(
    privileged_groups=[{"Gender": 1}],
    unprivileged_groups=[{"Gender": 0}]
)
RW.fit(data_aif)
data_transf = RW.transform(data_aif)
```

```python
instance_weights = data_transf.instance_weights

clf_rw = LogisticRegression()
clf_rw.fit(X_train_scaled, y_train, sample_weight=instance_weights[:len(y_train)])

y_pred_rw = clf_rw.predict(X_test_scaled)
print("Accuracy (After Reweighing):", accuracy_score(y_test, y_pred_rw))
```
